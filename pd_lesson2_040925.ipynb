```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
## update row + insert row + update cell

df = pd.read_csv("example.csv")

df.drop(1, axis=0)  # delete row with index 1
df.drop(index=1)  
df.drop('a', axis=1)  # delete column 'a'
df.drop(columns=['a'])

display(df)
df.loc [2] = [1, 2, 3, 4]  # update row index 2
df.loc [4] = [0, 0, 0, 0]  # insert row with index 4
df.loc [4] = [0, 0, 0, 12] # update row index 4
df.iloc[3, 2] = 999   # update cell at row 3 col 2
df.loc [2, 'd'] = 777  # update cell at index 2 col 'd'
df.iat [3, 3] = 888    # update cell at row 3 col 3 -- faster than iloc
df.at  [3, 'd'] = 888    # update cell at index 3 col 'd' -- faster than loc

df
```

```python
df = pd.read_csv("tips.csv")

df = df.set_index('Payment ID')

# update index Sun5260 column smoker to 'Yes'
df.at['Sun5260', 'smoker'] = 'Yes'

# update row1 col2 column sex to 'Female'
df.iat[1, 2] = 'Female'

# drop column 'day'
df.drop(columns=['day'])

# print lines from index Sat2657 to Sat3880
df.loc['Sat2657': 'Sat3880']

# drop row where index is Sat1766
df.drop(index='Sat1766')  # check ==> .loc['Sat1766']

# update all value in index Sun5260 to be: 21.01	3.50	'Male'	'No'	'Sun'	'Dinner'	3	7.00	'Travis Walters'	6011812112971322
#                    like the values of Sun4458

df.loc['Sun5260'] = [21.01,	3.50, 'Male', 'No', 'Sun', 'Dinner', 3, 7.00, 'Travis Walters',6011812112971322]
df.loc['Sun5260'] = df.loc['Sun4458']

df
```

```python
## concat by column

# First DataFrame
df1 = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

# Second DataFrame
df2 = pd.DataFrame({
    "C": ["x", "y", "z"],
    "D": ["p", "q", "r"]
})

# Concatenate by column (axis=1)
result = pd.concat([df1, df2], axis=1)

print(result)
```
